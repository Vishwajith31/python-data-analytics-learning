```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
```

```python
df_hourly = pd.read_csv('weather_data.csv')
df_hourly.head()
```

```python
df_hourly['date_time'] = pd.to_datetime(df_hourly['date_time'])
df_hourly.set_index('date_time', inplace=True)
df_hourly.head()
```

```python
# 1. resample data to weekly points
df_weekly = df_hourly.resample('W').first() # keeps 'season' column
df_weekly.head()
```

```python
# 2. Assign number to each week
df_weekly['idx'] = range(len(df_weekly)) # 0 to length of the dataframe
```

```python
# try to predict the temperature based on the number - trend
predictors = [
    'idx'
]

X = sm.add_constant(df_weekly[predictors])
Y = df_weekly['temperature_celsius']
```

```python
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
sns.scatterplot(x=df_weekly['idx'], y=df_weekly['temperature_celsius'])
plt.show()
```

```python
# Model Seasonality
print(pd.unique(df_weekly['season']))
```

```python
predictors = [
    'idx',
    'season'
]

X = sm.add_constant(pd.get_dummies(df_weekly[predictors], 
                                  columns=['season'],
                                  drop_first=True,
                                  dtype=int))
Y = df_weekly['temperature_celsius']
```

```python
X.head()
```

```python
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
df_weekly.tail()
```

```python
week210 = [1, 210, 0, 0, 1]
results.predict(week210)
```

```python
# Calculate residuals: actual - predicted
y_pred = results.predict(X)
y_actual = Y
residuals = y_actual - y_pred
```

```python
residuals.head()
```

```python
sns.regplot(x=y_pred, y=y_actual, line_kws={'color':'black'})
plt.xlabel('Predicted Temperature')
plt.ylabel('Actual Temperature')
plt.grid(alpha=0.7, linestyle='--')
plt.show()
```

```python
residuals.max()
```

```python
MAE = residuals.abs().mean()
print(MAE)
```
