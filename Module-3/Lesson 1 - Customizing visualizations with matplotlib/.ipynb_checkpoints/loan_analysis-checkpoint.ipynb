```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('loan_data.csv')
```

```python
grouped_data = df.groupby(['grade', 'homeownership'])['loan_amount'].sum().unstack()
```

```python
# ax = grouped_data.plot(kind='bar', stacked=True, color=['lightgray', 'darkgray', 'mediumseagreen'])
# plt.title('Total Loan Amount by Grade and Home Ownership', fontsize=12, pad=15, fontweight='bold')
# plt.xlabel('Grade', fontsize=11)
# plt.ylabel('Total Loan Amount ($)', fontsize=11)
# plt.grid(axis='y', linestyle='--', alpha=0.3)
# plt.xticks(rotation=0)
# plt.legend(title='Home Ownership')
# plt.show()
```

```python
ax = grouped_data.plot(kind='barh', stacked=True, color=['lightgray', 'darkgray', 'mediumseagreen'])
plt.title('Total Loan Amount by Grade and Home Ownership', fontsize=12, pad=15, fontweight='bold')
plt.ylabel('Grade', fontsize=11)  # Y-axis now shows Grade (switched)
plt.xlabel('Total Loan Amount ($)', fontsize=11)
plt.grid(axis='x', linestyle='--', alpha=0.3)  # Grid aligns with x-axis now
plt.yticks(rotation=0)
plt.legend(title='Home Ownership')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['annual_income'], df['total_credit_limit'], alpha=0.5, marker='.')
plt.title('Annual Income vs Total Credit Limit', fontsize=16, pad=15,)
plt.xlabel('Annual Income ($)', fontsize=14)
plt.ylabel('Total Credit Limit ($)', fontsize=14)
plt.xticks(rotation=0)
plt.xlim(0, 500000)
plt.show()
```

```python
# visualize credit limits of customers in top 5% of distribution for income
# find top 5% incomes
top_5_percent_income = df['annual_income'].quantile(0.95)
top_5_percent_income
```

```python
plt.scatter(df['annual_income'], df['total_credit_limit'], alpha=0.5, marker='.')
plt.title('Annual Income vs Total Credit Limit', fontsize=16, pad=15,)
plt.xlabel('Annual Income ($)', fontsize=14)
plt.ylabel('Total Credit Limit ($)', fontsize=14)
plt.xticks(rotation=0)
plt.xlim(0, 500000)
plt.axvline(x=top_5_percent_income, color='black', linestyle='--')
plt.show()
```
