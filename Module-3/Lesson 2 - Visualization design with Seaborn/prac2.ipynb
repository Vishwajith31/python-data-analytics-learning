```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
states = ['DC', 'AK', 'HI']
filtered_df = df[df['state'].isin(states)]
```

```python
grouped_data = filtered_df.groupby(['state', 'grade'])['loan_amount'].mean()
grouped_data.unstack().plot(kind='bar')
plt.show()
```

```python
sns.barplot(
    filtered_df,
    x='state',
    y='loan_amount',
    hue='grade'
)
plt.show()
```

```python
sns.barplot(
    filtered_df,
    x='state',
    y='loan_amount',
    hue='grade'
)
plt.title('Loan Amount by State and Grade')
plt.xlabel('State')
plt.ylabel('Loan Amount')
plt.legend(title='Grade', bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.barplot(
    filtered_df,
    x='state',
    y='interest_rate',
    hue='grade',
    estimator=np.max
)
plt.title('Loan Amount by State and Grade')
plt.xlabel('State')
plt.ylabel('Interest Rate')
plt.legend(title='Grade', bbox_to_anchor=(1,1))
plt.show()
```

```python

sns.barplot(
    filtered_df,
    x='state',
    y='loan_amount',
    hue='grade',
    palette='Blues'
)
plt.title('Loan Amount by State and Grade')
plt.xlabel('State')
plt.ylabel('Loan Amount')
plt.legend(title='Grade', bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.barplot(
    filtered_df,
    x='state',
    y='loan_amount',
    hue='grade',
    palette='icefire'
)
plt.title('Loan Amount by State and Grade')
plt.xlabel('State')
plt.ylabel('Loan Amount')
plt.legend(title='Grade', bbox_to_anchor=(1,1))
plt.show()
plt.savefig('loan_rates_graph.png')
```

```python
sns.boxplot(df, x='interest_rate')
plt.show()
```

```python
sns.boxplot(df, y='interest_rate')
plt.title('Distribution of Interest Rate', fontsize=16, fontweight='bold')
plt.ylabel('Interest Rate', fontsize=14)
sns.despine(bottom=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(df, hue='grade', y='interest_rate', palette='RdYlGn_r')
plt.title('Distribution of Interest Rate', fontsize=16, fontweight='bold')
plt.ylabel('Interest Rate', fontsize=14)
sns.despine()
plt.grid(axis='y', linestyle='--', color='black')
plt.show()
# plt.savefig()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(8,6))
sns.histplot(df, x='loan_amount', color='RosyBrown', bins=30, binwidth=5000)
plt.title('Distribution of Loan Amount', fontsize=16, pad=15, fontweight='bold')
plt.xlabel('Loan Amount', fontsize=14)
plt.grid(axis='y', linestyle='--', color='black')
sns.despine()
plt.show()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(8,6))
sns.histplot(df, x='loan_amount', color='RosyBrown', bins=30, binwidth=5000, kde=True)
plt.title('Distribution of Loan Amount', fontsize=16, pad=15, fontweight='bold')
plt.xlabel('Loan Amount', fontsize=14)
plt.grid(axis='y', linestyle='--', color='black')
sns.despine()
plt.show()
```

```python
correlation_features = [
    'emp_length', 'debt_to_income', 'annual_income', 'open_credit_lines', 'total_credit_limit', 'total_credit_utilized', 'total_debit_limit', 'loan_amount', 
    'interest_rate', 'installment', 'balance', 'paid_interest'
]

correlation_table = df[correlation_features].corr()

custom_cmap = sns.diverging_palette(240, 10, as_cmap=True)

plt.figure(figsize=(8,8))
sns.heatmap(
    correlation_table,
    annot=True,
    fmt='.2f',
    cmap=custom_cmap,
    center=0,
    vmin=-1,
    vmax=1,
    cbar_kws={'shrink': .8}
)
plt.title('Correlation Heatmap of Loan Features')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(
    df, 
    hue='grade',
    y='loan_amount',
    inner='box',
    palette='RdYlGn_r'
)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Distribution of Loan Amounts by Loan Grade with Internal Box')
plt.xlabel('Loan Grade')
plt.ylabel('Loan Amount')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    df,
    x='loan_amount',
    y='paid_interest',
    scatter_kws={'color': 'slateblue'},
    line_kws={'color': 'black'}
)
plt.title('Regression of Paid Interest on Loan Amount', fontsize=16, fontweight='bold')
plt.xlabel('Loan Amount ($)')
plt.ylabel('Paid Interest ($)')
plt.grid(linestyle='--', linewidth='0.5', color='gray')
plt.xlim(left=0)
plt.show()
```
